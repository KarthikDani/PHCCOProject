```python
import cobra

# Load a COBRA model (replace 'model.xml' with your actual model file)
model = cobra.io.read_sbml_model('/Users/karthik/Desktop/PHCCO IISc Internship/EMT/MODEL1602080000_url.xml')

# Dictionary to store reaction groups by subsystem
subsystem_groups = {}

# Populate the subsystem_groups dictionary
for reaction in model.reactions:
    subsystem = reaction.subsystem
    if subsystem not in subsystem_groups:
        subsystem_groups[subsystem] = []
    subsystem_groups[subsystem].append(reaction.id)

# Print all reaction groups
for subsystem, reactions in subsystem_groups.items():
    print(f"Subsystem: {subsystem}")
    print(f"Reactions: {', '.join(reactions)}")
    print()

# Example: Print all metabolite groups based on compartments
compartment_groups = {}
for metabolite in model.metabolites:
    compartment = metabolite.compartment
    if compartment not in compartment_groups:
        compartment_groups[compartment] = []
    compartment_groups[compartment].append(metabolite.id)

# Print all metabolite groups
for compartment, metabolites in compartment_groups.items():
    print(f"Compartment: {compartment}")
    print(f"Metabolites: {', '.join(metabolites)}")
    print()

# Example: Print all gene groups based on gene associations
gene_groups = {}
for reaction in model.reactions:
    for gene in reaction.genes:
        gene_id = gene.id
        if gene_id not in gene_groups:
            gene_groups[gene_id] = []
        gene_groups[gene_id].append(reaction.id)

# Print all gene groups
for gene_id, reactions in gene_groups.items():
    print(f"Gene: {gene_id}")
    print(f"Reactions: {', '.join(reactions)}")
    print()
```
